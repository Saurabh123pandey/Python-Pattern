```python
n=5
for i in range(0,n):
    for j in range(0,n):
        if i==j or i==n-1 or j==0:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()

```

```python
n=6
for i in range(0,n):
    for j in range(0,n):
        if i==0 or j==0 or i+j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()

```

```python
n=5
for i in range(0,n+1):
    for j in range(0,n+1):
        if j <= n - i:
            print(" ",end=" ")
        else:
            print("*",end=" ")
    print()
```
